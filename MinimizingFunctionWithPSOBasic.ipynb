```python
#Import the library what we will use 
import random
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Parameter setting
W = 0.9 
c1 = 0.8 
c2 = 0.5  

n_iterations = 150
target_error = 0.001 
n_particles = 120 
target = 1
```

```python
#function that models the problem
def fitness_function(position):
    return ((position[0]-1) **2) + (100*((position[1]-(position[0]**2)) **2))
```

```python
#particle generation function 
particle_position_vector = np.array([np.array([np.random.uniform(-1.5,1.5),np.random.uniform(-0.50,2.50)]) for _ in range(n_particles)])
pbest_position = particle_position_vector
pbest_fitness_value = np.array([float('inf') for _ in range(n_particles)])
gbest_fitness_value = float('inf')
gbest_position = np.array([float('inf'), float('inf')])
velocity_vector = ([np.array([0, 0]) for _ in range(n_particles)])
```

```python
#Function that manipulates the particle so that it is in the solution space
def Check_function():
 for Check in range(n_particles):
  if(pbest_position[Check][0]+pbest_position[Check][1]-2>0 and ((pbest_position[Check][0]-1)**3)-pbest_position[Check][1]+1>0):
    print(pbest_position[Check][0]+pbest_position[Check][1]-2)
    pbest_position[Check][0]=np.random.uniform(-1.5,1.5)
    pbest_position[Check][1]=np.random.uniform(-0.50,2.50)
    Check_function()
  else:print("particle ready")
```

```python
#Function that manipulates the particle so that the vector is in the solution space
def CheckVector_function():
   for CheckVector in range(n_particles):
     if(particle_position_vector[CheckVector][0]+particle_position_vector[CheckVector][1]-1<0 and ((particle_position_vector[CheckVector][0]-1)**3)-(particle_position_vector[CheckVector][1]+1)<=0):
       print("particle ready")
     else:
       particle_position_vector[CheckVector][0]=np.random.uniform(-1.5,1.5)
       particle_position_vector[CheckVector][1]=np.random.uniform(-0.51,2.51)
       CheckVector_function()
```

```python
listx=[] #list for x-axis
listy=[]# list for y-axis
listbest=[] # list for the best
iteration = 0
while iteration < n_iterations:
#Comparing and replacing pbest and gbest (in space)
    for i in range(n_particles):
        if(particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0):
            fitness_cadidate = fitness_function(particle_position_vector[i])
            print(iteration, ' ', fitness_cadidate, ' ', particle_position_vector[i])
            if(particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0):
                if(pbest_fitness_value[i] > fitness_cadidate and particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0 and pbest_position[i][0]+pbest_position[i][1]-2<0 and ((pbest_position[i][0]-1)**3)-pbest_position[i][1]+1<0):
                 pbest_fitness_value[i] = fitness_cadidate
                 pbest_position[i] = particle_position_vector[i]
                # print(particle_position_vector[i])
                if(gbest_fitness_value > fitness_cadidate and particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0 and pbest_position[i][0]+pbest_position[i][1]-2<0 and ((pbest_position[i][0]-1)**3)-pbest_position[i][1]+1<0):
                 gbest_fitness_value = fitness_cadidate
                 gbest_position = particle_position_vector[i]
                if(gbest_fitness_value > fitness_cadidate and particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0 and pbest_position[i][0]+pbest_position[i][1]-2<0 and ((pbest_position[i][0]-1)**3)-pbest_position[i][1]+1<0):
                  gbest_fitness_value = fitness_cadidate
                  gbest_position = particle_position_vector[i]      
                if(abs(gbest_fitness_value - target) < target_error):
                   break

    #Adding an element to the list
    for xi in range(n_particles): 
      listx.append(pbest_position[xi][0])
      listy.append(pbest_position[xi][1])
      listbest.append(gbest_fitness_value)
      #new location and vector creation
    for i in range(n_particles):
       if(particle_position_vector[i][0]+particle_position_vector[i][1]-1<0 and ((particle_position_vector[i][0]-1)**3)-(particle_position_vector[i][1]+1)<=0 and pbest_position[i][0]+pbest_position[i][1]-2<0 and ((pbest_position[i][0]-1)**3)-pbest_position[i][1]+1<0):
            fitness_cadidate = fitness_function(particle_position_vector[i])
            new_velocity = (W*velocity_vector[i]) + (c1*random.random()) * (pbest_position[i] - particle_position_vector[i]) + (c2*random.random()) * (gbest_position-particle_position_vector[i])
            new_position = new_velocity + particle_position_vector[i]
            particle_position_vector[i] = new_position

#printing and image operations
    plt.plot(listx,listy,'ro')
    plt.title(iteration)
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.show()
    listx.clear()
    listy.clear()
    iteration = iteration + 1
    
print("The best position is ", gbest_position, "in iteration number ", iteration, "The best function is:", gbest_fitness_value)
```
